# **假設:**
# **抓取台灣OTC市場中的上市上櫃公司，在2020至2023年間的完整ESG(月頻率)資料**

1.確定有訪問此類數據的權限。

2.確認Refinitiv提供的API支持台灣OTC市場的數據。

3.定義時間範圍，並提取資料。

```python
# 導入所需的庫
import refinitiv.data as rd
from refinitiv.data.content import esg

# 開啟數據會話
rd.open_session('platform.rdp')  # 需要有效的API密鑰

# 定義台灣OTC市場上市上櫃公司的RIC代碼列表
otc_instruments = ['RIC1', 'RIC2', 'RIC3']  # 請用實際的RIC代碼替換

# 定義時間範圍
start_date = '2020-01-01'
end_date = '2023-12-31'

# 提取ESG數據
for instrument in otc_instruments:
    response = esg.full_measures.Definition(
        instrument,
        start=start_date,
        end=end_date,
        interval=esg.Intervals.MONTHLY
    ).get_data()
    print(f"Data for {instrument}:")
    print(response.data.df)

# 關閉數據會話
rd.close_session()
```

**下載成.csv檔**

```python
import pandas as pd

#使用to_csv方法將DataFrame保存為CSV檔案
response.data.df.to_csv("您的檔案名稱.csv", index=False)
```
